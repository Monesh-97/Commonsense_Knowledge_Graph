```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('bert-large-nli-cls-token')
```

```python
sentences = ['This framework generates embeddings for each input sentence',
    'Sentences are passed as a list of string.', 
    'The quick brown fox jumps over the lazy dog.']
sentence_embeddings = model.encode(sentences[0])
```

```python
sentence_embeddings
```

```python
bert_embs='output/embeddings/cskg_embeddings_bert_nli_large.txt.gz'
```

```python
import gzip
from tqdm import tqdm
```

```python
ready=[]
with gzip.open(bert_embs, 'rb') as f:
    header=next(f)
    for line in tqdm(f, total=4500000):
        #print(line)
        line=line.decode()
        node, prop, data=line.split('\t')
        if prop=='embedding_sentence': continue
            
        ready_row = ' '.join([node, *(data.split(','))])
        ready.append(ready_row)
```

```python
ready[0]
```

```python
dim=len(ready[0].split(' '))-1
```

```python
i=len(ready)
```

```python
shape=[i, dim]
print(shape)
```

```python
with open('output/embeddings/bert_nli_large_w2v_format.txt', 'w') as w:
    hdr='%d %d\n' % (i, dim)
    w.write(hdr)
    for item in ready:
        w.write(item)
```
