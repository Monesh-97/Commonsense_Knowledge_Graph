```python
import os
import csv
import numpy as np
import pandas as pd
from collections import defaultdict
```

```python
path_ = "output2.tsv"
path='.'
```

```python

def load_data(path):
    tsv = open(path)
    data = pd.read_csv(tsv, sep="\t")

    my_data = data[['image_id','region_id','subject', 'object', 'subject_synset', 'object_synset', 'sentence']].copy()
    return my_data

my_data = load_data(path_)
my_data.head(50)
```

```python
def get_phraseInfo():
    sub2phrase = defaultdict(list)
    phrase2info = {}
    for index, row in my_data.iterrows():

        sub = row['subject']
        ob = row['object']
        phrase = row['sentence']
        phrase2info[phrase]={"object":ob, "subject":sub}
        if phrase not in sub2phrase[sub]:
            sub2phrase[sub].append(phrase)
            
    return sub2phrase, phrase2info
```

```python
sub2phrase, phrase2info = get_phraseInfo()
```

```python


def process_sentences(s1, s2, obj):
#    ob_mappings = {}
#    sub2phrase = defaultdict(list)
#    phrase2info = {}
    obj = obj[2:-2]
    s1 = s1.strip(".").strip(" ").split(' ')
    s2 = s2.strip(".").strip(" ").split(' ')
    if obj in s2: 
        ind = s2.index(obj)
    elif obj+"s" in s2:
        ind = s2.index(obj+"s")
        
    else:
        ind = 0
    if s1[len(s1)-1] == obj:
        return ' '.join(s1)+" "+ ' '.join(s2[ind+1:]) 
#    else:
#        return ' '.join(s1)+" and "+' '.join(s2)  #for now joined with "and". Need to think of some way.
```

```python
with open(os.path.join(path,'generative.tsv'), 'wt') as out_file:
    tsv_writer = csv.writer(out_file, delimiter='\t')
    tsv_writer.writerow(['Sentence 1', 'Sentence 2', 'Combined'])
    for index, row in my_data.iterrows():
        obj = row['object']
        s1 = row['sentence']
        if obj in sub2phrase.keys():
            for s2 in sub2phrase[obj]:
                if phrase2info[s2]["object"]!=row['subject']:
                    tsv_writer.writerow([s1, s2, process_sentences(s1,s2,obj)])
```
